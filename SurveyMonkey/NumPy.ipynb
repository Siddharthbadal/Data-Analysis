```python
a = np.array([[1,2,3],[4,5,6]])
b = np.array([3,4,5])
print(a)
print(b)
print(np.sum([a,b]))
# so numpy consider both array of the same size 
#and perform the operation on both th row. so it gives another array not one number.
#this is called array broadcasting
```

```python
np.subtract(a,b)
```

```python
np.multiply(a,b)
```

```python
a = np.array([[1,2,3],[4,5,6]])
b = np.array([[3,4,5],[1,2,3]])
print(a)
print(b)
print(np.sum([a,b]))
```
