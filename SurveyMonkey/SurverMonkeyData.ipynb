```python
import pandas as pd
import os
```

```python
pwd =os.getcwd() 
```

```python
dataset = pd.read_excel(pwd +".\Data.xlsx", sheet_name='Edited_Data')
```

```python
dataset
```

```python
dataset_modified = dataset.copy()
dataset_modified.head()
```

```python
dataset_modified.columns
```

```python
columns_to_drop = ['Start Date ', 'End Date ', 'Email Address ',
       'First Name ', 'Last Name ', 'Custom Data 1 ']
```

```python
dataset_modified= dataset_modified.drop(columns=columns_to_drop)
```

```python
dataset_modified.columns
```

```python
dataset_modified.head()
```

```python
id_vars = list(dataset_modified.columns[0:8])
value_vars = list(dataset_modified.columns[8:])
```

```python
dataset_melted = dataset_modified.melt(id_vars=id_vars, value_vars=value_vars, var_name="Question + SubQuestion" , value_name="Aanswer")
dataset_melted.head()
```

```python
dataset_melted.shape
```

```python
questions_import = pd.read_excel(pwd +".\Data.xlsx", sheet_name='Question')
questions_import
```

```python
questions_import.columns
```

```python
questions = questions_import.copy()
questions.drop(columns=["Raw Question","Raw SubQuestions","SubQuestion"],inplace=True)
```

```python
questions.columns
```

```python
dataset_melted.columns
```

```python
dataset_merged = pd.merge(left=dataset_melted, right=questions, how="left", left_on="Question + SubQuestion", right_on="Question +  SubQuestion")
len(dataset_melted)
len(dataset_merged)
```

```python
dataset_merged.head()
```

```python
respondents = dataset_merged[dataset_merged['Aanswer'].notna()]
respondents
```

```python
respondents= respondents.groupby('Questions')['Respondent ID '].unique().reset_index()
respondents.rename(columns={'Respondent ID ':'Respondents'}, inplace=True)
respondents
```

```python
dataset_merged_two = pd.merge(left=dataset_merged, right=respondents, how="left", left_on="Questions", right_on="Questions")
print(len(dataset_melted))
print(len(dataset_merged_two))
dataset_merged_two
```

```python
same_answer = dataset_merged #[dataset_merged['Aanswer'].notna()]
same_answer= same_answer.groupby(['Question + SubQuestion', 'Aanswer'])['Respondent ID '].unique().reset_index()
same_answer.rename(columns={'Respondent ID ':'same_answer'}, inplace=True)
same_answer
```

```python
dataset_merged_third = pd.merge(left=dataset_merged_two, right=same_answer, how="left", left_on=['Question + SubQuestion', 'Aanswer'], right_on=['Question + SubQuestion', 'Aanswer'])
dataset_merged_third['same_answer'].fillna(0, inplace=True)
print(len(dataset_merged_two))
print(len(dataset_merged_third))
dataset_merged_third
```

```python
Output = dataset_merged_third.copy()
Output.rename(columns={"Identify which division you work in. Response":"Division Primary",
                       "Identify which division you work in. Other (please specify)":"Division Secondary",
                       "Which of the following best describes your position level? Response":"Position",	
                       "Which generation are you apart of? Response":"Generation",	
                       "Please select the gender in which you identify. Response":"Gender",
                       "Which duration range best aligns with your tenure at your company? Response":"Tenure",
                       "Which of the following best describes your employment type? Response":"Employment_type"}, inplace=True)
Output
```

```python
Output.to_excel(pwd + ".\\final_output.xlsx", index=False)
```
