## Udemy Courses Data

```python
import pandas as pd
```

```python
data  = pd.read_csv("Udemy_Courses.csv")
```

```python
data.head(5)
```

### Different kind of courses offered by Udemy

```python
data.subject.unique()
```

```python
#max courses on a subject
data['subject'].value_counts()
```

```python
#free courses

data[data.is_paid == False].head(2)
```

```python
# paid courses

data[data.is_paid == True].head(2)
```

### total courses by subject

```python

data.groupby('subject').count()
```

```python
# free courses as per subjectect category
data[data.is_paid==False].groupby('subject').count()
```

```python
# Paid courses as per subjectect category
data[data.is_paid==True].groupby('subject').count()
```

### Top courses

```python
# by number od subscribers
data.sort_values('num_subscribers', ascending = False).head(5)
```

```python
# by duration
data.sort_values('content_duration', ascending = False).head(5)
```

### all the web dev courses with prices below 100

```python
data[(data.subject == 'Web Development') & (data.price > '100')].head(5)
```

```python
data[(data.subject == 'Web Development') & (data.price == '100')].head(5)
```

```python
# Python courses
data[data.course_title.str.contains('Python')]
```

### Course published year

```python
data['published_timestamp'] = pd.to_datetime(data.published_timestamp)
data['year'] = data['published_timestamp'].dt.year
data.head()
```

```python
data[data.year == '2015'].head()
```
