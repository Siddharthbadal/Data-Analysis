### find the null value and fill  these values

```python
#car.isnull()
car.isnull().sum()
```

```python
car.fillna(car.mean(), inplace = True)
```

```python
# column Make records total
car['Make'].value_counts()
```

```python
#Where origin is Asia or Europe
car[car['Origin'].isin(['Asia','Europe'])]
```

```python
#remove the unwanted recodes: where weigth is above 5000

car[~(car['Weight'] > 5000)]
```

```python
# increase all the values of MPG_City by 

car['MPG_City'] = car['MPG_City'].apply(lambda x:x*2)
```
