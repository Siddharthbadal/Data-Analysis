```python
import seaborn as sns
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
```

```python
#number of dinners and lunches (time)
tips.time.value_counts()
```

```python
#plot the time
tips.time.value_counts().plot(kind='bar')
```

```python
#proprtion in percantages
tips.time.value_counts(normalize=True)
```

```python
#proprtion in percantages sex
tips.sex.value_counts(normalize=True)
```

```python
#when does people more tips. This is one dimensional with one column
tips.groupby(['time'])['tip'].mean()
```

```python
#who payes more tips male or females
tips.groupby(['sex'])['tip'].mean()
```

```python
#whic day has more tip
tips.groupby(['day'])['tip'].mean()
```

```python
#plot it 
tips.groupby(['time'])['tip'].mean().plot.bar()
```

```python
#tips by smokers
tips.groupby(['smoker'])['tip'].mean()
```

```python
# cross table with two dimensional. sex and smoker
tips.groupby(['smoker','sex'])['tip'].mean()
```

```python
# a table for above
tips.groupby(['smoker','sex'])['tip'].mean().unstack()
```

```python
tips.groupby(['total_bill','size'])['tip'].mean().head()
```

```python
# perceantage of tip of total bill
tips['tip_pct %'] = round((tips.tip / tips.total_bill)*100,2)
tips.head()
```

```python
# visualize the relationship between toal bill and tips
tips.plot.scatter(x='total_bill', y='tip')
```

```python
# visualize the relationship between toal bill and tips using sns
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=tips)
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='time')
```

```python
#birfurcation of time on the chart based on column sex
sns.relplot(x="total_bill", y='tip', data=tips, col='sex', hue='time')
```

```python
#distribution of bill by time
sns.displot(data=tips, x="total_bill", col="time", kde=True)
```
